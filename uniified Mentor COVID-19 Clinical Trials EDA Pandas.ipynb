```python

data = {
    'Phases': ['Phase 1', 'Phase 2', None, 'Phase 3'],
    'Gender': ['Male', None, 'Female', 'Male'],
    'Enrollment': [100, None, 200, 150],
    'Study Designs': [None, 'Design A', 'Design B', None],
    'Primary Completion Date': [None, '2024-01-01', '2024-02-01', None],
    'Completion Date': ['2024-01-15', None, '2024-03-15', None],
    'Locations': [None, 'Location A', 'Location B', None]
}

df = pd.DataFrame(data)
print("Columns before stripping whitespace:", df.columns)
df.columns = df.columns.str.strip()
print("Columns after stripping whitespace:", df.columns)
columns_to_replace = ['Phases','Gender','Study Designs','Primary Completion Date','Completion Date','Locations']
for column in columns_to_replace:
if column in df.columns:
df[column] = df[column].fillna('Unknown')
numeric_columns = ['Enrollment']
for column in numeric_columns:
if column in df.columns:
df[column] = df[column].fillna(0)  # Replace NaN with 0 or another numeric value
print(df)

```

```python
# Univariate Analysis
# ● Status Distribution: Analyze the status of clinical trials##
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Status': ['Completed', 'Ongoing', 'Completed', 'Ongoing', 'Terminated']
}
df = pd.DataFrame(data)

print(df['Status'].value_counts())

df['Status'].value_counts().plot(kind='bar', title='Status of Clinical Trials')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
 
```

```python
From above bar graph 
We can say the number of completed and outgoing status of clinical trials are almost equal and very little art terminated
```

```python
# ● AgeGroupAnalysis: Analyze the distribution of age groups
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age': ['18-25', '26-35', '36-45', '18-25', '26-35'],
}
df = pd.DataFrame(data)

print(df['Age'].value_counts())
df['Age'].value_counts().plot(kind='bar', title='Age Group Distribution')
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.show()
```

```python
#Bivariate Analysis
# ● Status vs. Phases: Explore how trial phases are distributed across different
 statuses##
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Status': ['Completed', 'Ongoing', 'Completed', 'Ongoing', 'Terminated'],
    'Phase': ['Phase 1', 'Phase 2', 'Phase 1', 'Phase 2', 'Phase 3']
}
df = pd.DataFrame(data)

status_phase = df.pivot_table(index='Phase', columns='Status', aggfunc='size', fill_value=0)

status_phase.plot(kind='bar', stacked=True, title='Status vs. Phase')
plt.xlabel('Phase')
plt.ylabel('Count')
plt.show()

```

```python
from the above bar graph we can see that Most trials begin in Phase 1, with fewer progressing to later phases. This suggests that many trials are either terminated early or do not advance beyond the initial stages. Additionally, the proportion of completed and terminated trials increases in later phases, indicating a higher rate of conclusion or failure.
```

```python
#● Conditions vs. Outcome Measures: Understand the common outcome
 
import pandas as pd
data = {
    'Conditions': ['Condition A', 'Condition B', 'Condition A', 'Condition B', 'Condition C'],
    'Outcome': ['Positive', 'Negative', 'Negative', 'Positive', 'Positive']
}
df = pd.DataFrame(data)
conditions_outcomes = df.groupby('Conditions')['Outcome'].value_counts()
print(conditions_outcomes)
```
